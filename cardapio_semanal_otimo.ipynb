```python
import random

cardapio = [
    {
        "ingredientes": ["Pão de sal", "Ovo frito", "Suco de Goiaba"],
        "preco": 3.12,
        "calorias": 348,
        "proteinas": 12.68,
        "carboidratos": 51.39,
        "fibras": 1.3,
        "refeicao": "Café da manhã",
        "escolhas": 0 # Guarda a quantidade de vezes que a receita foi escolhida na semana
    },
    {
        "ingredientes": ["Cuscuz", "Margarina", "Café com Leite"],
        "preco": 1.69,
        "calorias": 280,
        "proteinas": 5.6,
        "carboidratos": 43.5,
        "fibras": 2.05,
        "refeicao": "Café da manhã",
        "escolhas": 0
    },
    {
        "ingredientes": ["Mamão", "Uva", "Yorgute"],
        "preco": 5.61,
        "calorias": 207,
        "proteinas": 9.79,
        "carboidratos": 28.9,
        "fibras": 2.73,
        "refeicao": "Café da manhã",
        "escolhas": 0
    },
    {
        "ingredientes": ["Tapioca com Queijo Coalho", "Vitamina de Banana"],
        "preco": 3.90,
        "calorias": 320,
        "proteinas": 10.5,
        "carboidratos": 45.2,
        "fibras": 3.1,
        "refeicao": "Café da manhã",
        "escolhas": 0
    },
    {
        "ingredientes": ["Pão Integral", "Queijo Minas", "Geleia de Frutas"],
        "preco": 2.98,
        "calorias": 295,
        "proteinas": 14.2,
        "carboidratos": 38.7,
        "fibras": 5.4,
        "refeicao": "Café da manhã",
        "escolhas": 0
    },
    {
        "ingredientes": ["frango assado", "Arroz", "Salada"],
        "preco": 5.49,
        "calorias": 395.5,
        "proteinas": 50.23,
        "carboidratos": 32.39,
        "fibras": 2.17,
        "refeicao": "Almoço",
        "escolhas": 0
    },
    {
        "ingredientes": ["Figado frito", "Arroz", "Feijão"],
        "preco": 2.50,
        "calorias": 532,
        "proteinas": 49.75,
        "carboidratos": 49.75,
        "fibras": 8.97,
        "refeicao": "Almoço",
        "escolhas": 0
    },
    {
        "ingredientes": ["Bisteca Suína", "Arroz", "Macarrão Alho e Óleo"],
        "preco": 3.74,
        "calorias": 603,
        "proteinas": 43.5,
        "carboidratos": 56.1,
        "fibras": 2.1,
        "refeicao": "Almoço",
        "escolhas": 0
    },
    {
        "ingredientes": ["Peixe Grelhado", "Purê de Batata", "Brócolis"],
        "preco": 8.25,
        "calorias": 485,
        "proteinas": 38.2,
        "carboidratos": 40.8,
        "fibras": 6.3,
        "refeicao": "Almoço",
        "escolhas": 0
    },
    {
        "ingredientes": ["Carne Moída", "Arroz Integral", "Lentilha", "Abobrinha Grelhada"],
        "preco": 5.19,
        "calorias": 520,
        "proteinas": 35.8,
        "carboidratos": 55.4,
        "fibras": 10.2,
        "refeicao": "Almoço",
        "escolhas": 0
    },
    {
        "ingredientes": ["Arroz", "Feijão", "Ovo cozido"],
        "preco": 1.42,
        "calorias": 282,
        "proteinas": 13.4,
        "carboidratos": 42.7,
        "fibras": 8.8,
        "refeicao": "Jantar",
        "escolhas": 0
    },
    {
        "ingredientes": ["Arroz", "Alface", "Bife Frito"],
        "preco": 5.13,
        "calorias": 533,
        "proteinas": 41.5,
        "carboidratos": 29.1,
        "fibras": 1.2,
        "refeicao": "Jantar",
        "escolhas": 0
    },
    {
        "ingredientes": ["Arroz", "Batata frita", "Frango Grelhado"],
        "preco": 4.15,
        "calorias": 631.5,
        "proteinas": 41.1,
        "carboidratos": 60.1,
        "fibras": 3.8,
        "refeicao": "Jantar",
        "escolhas": 0
    },
    {
        "ingredientes": ["Sopa de Legumes", "Frango Desfiado"],
        "preco": 2.52,
        "calorias": 320,
        "proteinas": 25.4,
        "carboidratos": 35.6,
        "fibras": 7.9,
        "refeicao": "Jantar",
        "escolhas": 0
    },
    {
        "ingredientes": ["Salada de Quinoa", "Atum", "Tomate Seco"],
        "preco": 7.05,
        "calorias": 385,
        "proteinas": 30.2,
        "carboidratos": 42.3,
        "fibras": 8.5,
        "refeicao": "Jantar",
        "escolhas": 0
    },
    {
        "ingredientes": ["Crepioca", "Leite desnatado"],
        "preco": 2.40,
        "calorias": 269,
        "proteinas": 17.5,
        "carboidratos": 28.6,
        "fibras": 0,
        "refeicao": "Ceia",
        "escolhas": 0
    },
    {
        "ingredientes": ["Cereal", "Leite Integral"],
        "preco": 2.32,
        "calorias": 321,
        "proteinas": 10.4,
        "carboidratos": 47.8,
        "fibras": 2.5,
        "refeicao": "Ceia",
        "escolhas": 0
    },
    {
        "ingredientes": ["Mingau de Aveia"],
        "preco": 1.52,
        "calorias": 271,
        "proteinas": 10.3,
        "carboidratos": 44.8,
        "fibras": 5.8,
        "refeicao": "Ceia",
        "escolhas": 0
    },
    {
        "ingredientes": ["Iogurte Grego", "Granola", "Mel"],
        "preco": 6.12,
        "calorias": 285,
        "proteinas": 15.8,
        "carboidratos": 35.2,
        "fibras": 3.2,
        "refeicao": "Ceia",
        "escolhas": 0
    },
    {
        "ingredientes": ["Vitamina de Abacate", "Castanhas"],
        "preco": 3.00,
        "calorias": 340,
        "proteinas": 8.7,
        "carboidratos": 28.5,
        "fibras": 7.1,
        "refeicao": "Ceia",
        "escolhas": 0
    }
]

# Funcao para calcular o custo total e os valores nutricionais de um cardapio semanal
def avaliar_cardapio(cardapio_semanal, cardapio_completo, metas_diarias, penalizacao_variedade):
    custo_total = 0
    nutricionais_semanais = {
        "calorias": 0,
        "proteinas": 0,
        "carboidratos": 0,
        "fibras": 0
    }
    refeicoes_escolhidas_contagem = {}

    for dia in cardapio_semanal:
        custo_dia = 0
        nutricionais_dia = {
            "calorias": 0,
            "proteinas": 0,
            "carboidratos": 0,
            "fibras": 0
        }
        refeicoes_escolhidas_dia = set()

        for refeicao_info in dia:
            for r in cardapio_completo:
                if r["ingredientes"] == refeicao_info["ingredientes"]:
                    refeicao_original = r
                    break

            porcao = refeicao_info["porcao"]

            custo_dia += refeicao_original["preco"] * porcao
            nutricionais_dia["calorias"] += refeicao_original["calorias"] * porcao
            nutricionais_dia["proteinas"] += refeicao_original["proteinas"] * porcao
            nutricionais_dia["carboidratos"] += refeicao_original["carboidratos"] * porcao
            nutricionais_dia["fibras"] += refeicao_original["fibras"] * porcao
            refeicoes_escolhidas_dia.add(tuple(refeicao_original["ingredientes"]))

        custo_total += custo_dia
        for nutriente in nutricionais_semanais:
            nutricionais_semanais[nutriente] += nutricionais_dia[nutriente]

        for refeicao in refeicoes_escolhidas_dia:
            refeicoes_escolhidas_contagem[refeicao] = refeicoes_escolhidas_contagem.get(refeicao, 0) + 1

    # Penalidade por falta de variedade (mesma refeicao escolhida mais de 3 vezes na semana)
    penalidade_variedade_total = 0
    for contagem in refeicoes_escolhidas_contagem.values():
        if contagem > 3:
            penalidade_variedade_total += (contagem - 3) * penalizacao_variedade

    custo_total += penalidade_variedade_total

    # Penalidade por nao atender as metas nutricionais (apenas para metas minimas)
    penalidade_nutricional = 0
    if nutricionais_semanais["proteinas"] < metas_diarias["proteinas"] * 7:
        penalidade_nutricional += (metas_diarias["proteinas"] * 7 - nutricionais_semanais["proteinas"]) * 10
    if nutricionais_semanais["carboidratos"] < metas_diarias["carboidratos"] * 7:
        penalidade_nutricional += (metas_diarias["carboidratos"] * 7 - nutricionais_semanais["carboidratos"]) * 5
    if nutricionais_semanais["fibras"] < metas_diarias["fibras"] * 7:
        penalidade_nutricional += (metas_diarias["fibras"] * 7 - nutricionais_semanais["fibras"]) * 20
    if nutricionais_semanais["calorias"] > metas_diarias["calorias"] * 7:
         penalidade_nutricional += (nutricionais_semanais["calorias"] - metas_diarias["calorias"] * 7) * 0.5


    custo_total += penalidade_nutricional

    return custo_total, nutricionais_semanais

# Funcao para gerar um cardapio semanal aleatorio
def gerar_cardapio_aleatorio(cardapio_completo, tipo_refeicoes):
    cardapio_semanal = []
    for _ in range(7): # Para cada dia da semana
        cardapio_dia = []
        for tipo in tipo_refeicoes:
            refeicoes_disponiveis = [r for r in cardapio_completo if r["refeicao"] == tipo]
            refeicao_escolhida = random.choice(refeicoes_disponiveis)
            porcao = random.uniform(1, 3)  # Porcao aleatoria entre 1 e 3
            cardapio_dia.append({"ingredientes": refeicao_escolhida["ingredientes"], "porcao": porcao})
        cardapio_semanal.append(cardapio_dia)
    return cardapio_semanal

# Funcao para realizar o cruzamento de dois cardapios semanais (individuos)
def cruzar_cardapios(cardapio1, cardapio2):
    ponto_corte = random.randint(1, 6) # Ponto de corte aleatorio entre os dias da semana
    novo_cardapio1 = cardapio1[:ponto_corte] + cardapio2[ponto_corte:]
    novo_cardapio2 = cardapio2[:ponto_corte] + cardapio1[ponto_corte:]
    return novo_cardapio1, novo_cardapio2

# Funcao para aplicar mutacao em um cardapio semanal
def mutar_cardapio(cardapio_semanal, cardapio_completo, tipo_refeicoes, taxa_mutacao=0.1):
    for dia in cardapio_semanal:
        for i in range(len(dia)):
            if random.random() < taxa_mutacao:
                tipo = [r["refeicao"] for r in cardapio_completo if r["ingredientes"] == dia[i]["ingredientes"]][0]
                refeicoes_disponiveis = [r for r in cardapio_completo if r["refeicao"] == tipo]
                nova_refeicao = random.choice(refeicoes_disponiveis)
                dia[i]["ingredientes"] = nova_refeicao["ingredientes"]
                dia[i]["porcao"] = random.uniform(1, 3) # Mutacao na porcao tambem
    return cardapio_semanal

# Implementacao do Algoritmo Genetico
def algoritmo_genetico(cardapio_completo, metas_diarias, tipo_refeicoes, tamanho_populacao=100, geracoes=100, taxa_mutacao=0.1, penalizacao_variedade=500):
    populacao = [gerar_cardapio_aleatorio(cardapio_completo, tipo_refeicoes) for _ in range(tamanho_populacao)]

    for geracao in range(geracoes):
        # Avaliar a populacao
        avaliacoes = [(avaliar_cardapio(individuo, cardapio_completo, metas_diarias, penalizacao_variedade), individuo) for individuo in populacao]
        avaliacoes.sort(key=lambda x: x[0][0]) # Ordenar pelo custo (fitness)

        # Selecionar os melhores individuos (elitismo)
        elite_size = int(tamanho_populacao * 0.1)
        nova_populacao = [individuo for avaliacao, individuo in avaliacoes[:elite_size]]

        # Gerar novos individuos por cruzamento e mutacao
        while len(nova_populacao) < tamanho_populacao:
            pai1 = random.choice(avaliacoes[:int(tamanho_populacao * 0.5)])[1] # Selecionar pais da metade superior
            pai2 = random.choice(avaliacoes[:int(tamanho_populacao * 0.5)])[1]
            filho1, filho2 = cruzar_cardapios(pai1, pai2)
            nova_populacao.append(mutar_cardapio(filho1, cardapio_completo, tipo_refeicoes, taxa_mutacao))
            if len(nova_populacao) < tamanho_populacao:
                nova_populacao.append(mutar_cardapio(filho2, cardapio_completo, tipo_refeicoes, taxa_mutacao))

        populacao = nova_populacao

    # Retornar o melhor individuo da ultima geracao
    melhor_cardapio = avaliacoes[0][1]
    melhor_custo, melhores_nutricionais = avaliacoes[0][0]
    return melhor_cardapio, melhor_custo, melhores_nutricionais

# Executar o algoritmo genetico
melhor_cardapio_semanal, custo_final, nutricionais_finais = algoritmo_genetico(cardapio, metas_diarias, tipo_refeicoes)

# Exibir resultados do algoritmo genetico
print("\n" + "="*50)
print("  CARDAPIO OTIMIZADO (Algoritmo Genetico)")
print("="*50)

mapa_refeicoes = {tuple(r["ingredientes"]): r for r in cardapio}
custo_total_semana_ag = 0

for idx, dia_info in enumerate(melhor_cardapio_semanal):
    print(f"\n{'-' * 48} Dia {idx+1} {'-' * 48}")

    custo_dia = 0
    calorias_dia, proteinas_dia, carboidratos_dia, fibras_dia = 0, 0, 0, 0

    # Organizar as refeicoes do dia pela ordem definida em tipo_refeicoes
    dia_organizado = sorted(dia_info, key=lambda x: tipo_refeicoes.index([r["refeicao"] for r in cardapio if r["ingredientes"] == x["ingredientes"]][0]))

    for refeicao_info in dia_organizado:
        ingredientes_escolhidos = refeicao_info["ingredientes"]
        porcao = refeicao_info["porcao"]

        refeicao_original = mapa_refeicoes[tuple(ingredientes_escolhidos)]
        tipo_refeicao = refeicao_original["refeicao"]


        print(f"  -> {tipo_refeicao:<13} | {', '.join(ingredientes_escolhidos):<65} | Porção: {porcao:.2f}")

        custo_dia += refeicao_original['preco'] * porcao
        calorias_dia += refeicao_original['calorias'] * porcao
        proteinas_dia += refeicao_original['proteinas'] * porcao
        carboidratos_dia += refeicao_original['carboidratos'] * porcao
        fibras_dia += refeicao_original['fibras'] * porcao

    print("-" * 25)
    print(f"  Resumo do Dia:")
    print(f"    Custo       : R$ {custo_dia:.2f}")
    print(f"    Calorias    : {calorias_dia:.2f} kcal (Meta Semanal: <= {metas_diarias['calorias']*7})")
    print(f"    Proteínas   : {proteinas_dia:.2f} g (Meta Semanal: >= {metas_diarias['proteinas']*7})")
    print(f"    Carboidratos: {carboidratos_dia:.2f} g (Meta Semanal: >= {metas_diarias['carboidratos']*7})")
    print(f"    Fibras      : {fibras_dia:.2f} g (Meta Semanal: >= {metas_diarias['fibras']*7})")

    custo_total_semana_ag += custo_dia

print("\n" + "="*50)
print(f"  CUSTO TOTAL DA SEMANA (Algoritmo Genetico): R$ {custo_total_semana_ag:.2f}")
print("="*50)

print("\nResumo Nutricional Semanal (Algoritmo Genetico):")
print(f"  Calorias    : {nutricionais_finais['calorias']:.2f} kcal (Meta Semanal: <= {metas_diarias['calorias']*7})")
print(f"  Proteínas   : {nutricionais_finais['proteinas']:.2f} g (Meta Semanal: >= {metas_diarias['proteinas']*7})")
print(f"  Carboidratos: {nutricionais_finais['carboidratos']:.2f} g (Meta Semanal: >= {metas_diarias['carboidratos']*7})")
print(f"  Fibras      : {nutricionais_finais['fibras']:.2f} g (Meta Semanal: >= {metas_diarias['fibras']*7})")
```
